```python
df_global=pd.read_csv('/kaggle/input/project1/global_data.csv')
df_global.head()
```

```python
df_local=pd.read_csv('/kaggle/input/project1/local_data.csv')
df_local.head()
```

```python
import matplotlib.pyplot as plt
local_year = df_local['year'].values.tolist()
print (local_year)
local_temp=df_local['avg_temp'].values.tolist()
print (local_temp)
```

```python
global_year = df_global['year'].values.tolist()
print (global_year)
global_temp=df_global['avg_temp'].values.tolist()
print (global_temp)
```

```python
plt.plot(global_year, global_temp, color='red', marker='o', label='global avarage temperature')
plt.plot(local_year, local_temp, color='green', marker='o', label='local avarage temperature')
plt.title('Avarage Temperature Vs Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Avarage Temprature', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```
